#Step 4: Join the Grandelectors and Population Tables 

```python
#joining the grand electors and population tables on  the state column 
# we will need to create a new table which is what we will be using for analysis 
%%sql 
CREATE TABLE Elections AS 
SELECT Population.State, Population.population, GrandElectors.GrandElectors
FROM Population 
INNER JOIN GrandElectors
ON GrandElectors.State=Population.State;
```

```python
#preview the table 
%%sql 
SELECT * FROM Elections;
```

```python
#change the name "District of Columbia" to DC in the new table 
%%sql 
UPDATE Elections
SET State ="DC"
WHERE State="DISTRICT OF COLUMBIA";
```

```python
#preview the change 
%%sql 
SELECT * FROM Elections;
```

#Step 8: Get the priority list by ordering the states by decreasing ratio of grand electors per capita 

```python
#Order the states by decreasing ratio of Grand Electors per capita 
#priority list
%%sql 
SELECT * FROM Elections
ORDER BY Ratio DESC;
```

#Step 9: Compute the running total of Grandelectors 

```python
#compute the running total of the grandelectors in the sorted list 
%%sql
SELECT 
t.State, t. Grandelectors, t.Population, t.Ratio,
  (SELECT SUM(r.GrandElectors) FROM Elections as r WHERE r.Ratio <=t.Ratio) AS Running_total  
FROM Elections AS t
GROUP BY t.Ratio;
```

```python
%%sql 
CREATE TABLE US_elections AS 
SELECT 
t.State, t. GrandElectors, t.Population, t.Ratio,
  (SELECT SUM(r.GrandElectors) FROM Elections as r WHERE r.Ratio<=t.Ratio) AS Running_total  
FROM Elections AS t
GROUP BY t.Ratio ORDER BY t.Ratio;
```

```python
%%sql
SELECT * FROM US_elections;
```

#Step 10: Compute the half of the total Grand electors 

```python
#compute the half of the total Grand Electors overrall in the whole country 
#This is the threshold we need to reach for winning the presidential election.
%%sql 
SELECT SUM(GrandElectors)/2 
FROM Elections; 
```

#Step 11: Filter sorted list of states 

```python
#filter out the sorted list of states 
#keep only the top ones enabling us to reach the threshold 
#Hint: You can do that in 2 steps:
#Select all the states for which the running total is below or equal to the threshold.
#Add the first state for which the running total is larger than the threshold.
%%sql
SELECT * FROM Us_elections 
WHERE Running_total <= 269;
```

```python
#Add the first state for which the running total is larger than the threshold.
%%sql
SELECT * FROM US_elections 
WHERE Running_total>=269 LIMIT 1;
```
